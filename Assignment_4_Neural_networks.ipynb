```python
import numpy as np
```

```python
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

Y = np.array([
    [0],
    [1],
    [1],
    [0]
])
```

```python
def relu(z):
    return np.maximum(0,z)

def relu_derivative(z):
    return (z > 0).astype(float)
```

```python
np.random.seed(42)

W1 = np.random.randn(2,4)
B1 = np.zeros((1,4))

W2 = np.random.randn(4,1)
B2 = np.zeros((1,1))
```

```python
learning_rate = 0.1

for epoch in range(10000):

    # Forward Pass

    Z1 = X @ W1 + B1
    A1 = relu(Z1)

    Z2 = A1 @ W2 + B2
    A2 = relu(Z2)

    # Loss

    loss = np.mean((A2 - Y)**2)

    # Backward Pass

    dA2 = (A2 - Y)

    dZ2 = dA2 * relu_derivative(Z2)

    dW2 = A1.T @ dZ2
    dB2 = np.sum(dZ2, axis=0, keepdims=True)

    dA1 = dZ2 @ W2.T

    dZ1 = dA1 * relu_derivative(Z1)

    dW1 = X.T @ dZ1
    dB1 = np.sum(dZ1, axis=0, keepdims=True)

    # Gradient Descent Update

    W2 -= learning_rate*dW2
    B2 -= learning_rate*dB2

    W1 -= learning_rate*dW1
    B1 -= learning_rate*dB1

    if epoch % 1000 == 0:
        print("Epoch:", epoch, "Loss:", loss)
```

```python
Z1 = X @ W1 + B1
A1 = relu(Z1)

Z2 = A1 @ W2 + B2
A2 = relu(Z2)

print("Predictions:")
print(np.round(A2))
```

Modification Chosen:
I replaced the sigmoid activation function in the hidden layer with ReLU.

Mathematical Change:
ReLU computes max(0, z), whereas sigmoid computes 1/(1+e^(-z)). ReLU is piecewise linear and avoids saturation for positive inputs.

Expected Behaviour:
I expected faster learning because ReLU produces stronger gradients for positive activations. This helps reduce the vanishing gradient problem and often improves training efficiency.
